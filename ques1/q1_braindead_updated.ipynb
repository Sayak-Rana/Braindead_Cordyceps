```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Placement_Data_Full_Class.csv")
```

```python
df.head()
```

```python
# Checking for missing values
df.isnull().sum()

# Descriptive Statistics
df.describe()
```

# **What are the factors affecting the salary of a student?**

```python
# Correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Box plot of salaries by degree branch
sns.boxplot(x="degree_t", y="salary", data=df)
plt.title("Salary by degree branch")
plt.xlabel("degree_t")
plt.ylabel("Salary")
plt.show()
```

```python
# Scatter plot of salaries by degree prcentage
plt.scatter(x="degree_p", y="salary", data=df)
plt.title("Salary by degree percentage")
plt.xlabel("degree_p")
plt.ylabel("Salary")
plt.show()
```

```python
# Box plot of salaries by ssc board
sns.boxplot(x="ssc_b", y="salary", data=df)
plt.title("Salary by ssc board")
plt.xlabel("ssc_b")
plt.ylabel("Salary")
plt.show()
```

```python
# Scatter plot of salaries by ssc percentage
plt.scatter(x="ssc_p", y="salary", data=df)
plt.title("Salary by ssc percentage")
plt.xlabel("ssc_p")
plt.ylabel("Salary")
plt.show()
```

```python
# Box plot of salaries by hsc board
sns.boxplot(x="hsc_b", y="salary", data=df)
plt.title("Salary by hsc board")
plt.xlabel("hsc_b")
plt.ylabel("Salary")
plt.show()
```

```python
# Box plot of salaries by hsc stream
sns.boxplot(x="hsc_s", y="salary", data=df)
plt.title("Salary by hsc stream")
plt.xlabel("hsc_s")
plt.ylabel("Salary")
plt.show()
```

```python
# Box plot of salaries by hsc percentage
sns.boxplot(x="ssc_b", y="salary", data=df)
plt.title("Salary by hsc percentage")
plt.xlabel("hsc_p")
plt.ylabel("Salary")
plt.show()
```

```python
# Scatter plot of salaries by mba percentage
plt.scatter(x="mba_p", y="salary", data=df)
plt.title("Salary by mba percentage")
plt.xlabel("mba_p")
plt.ylabel("Salary")
plt.show()
```

```python
# Box plot of salaries by etest percentage
plt.scatter(x="etest_p", y="salary", data=df)
plt.title("Salary by etest percentage")
plt.xlabel("etest_p")
plt.ylabel("Salary")
plt.show()
```

# **Which factors affect placement?**

###### Males have a higher chance of placement

```python
# Bar plot of placement and gender
counts = df.groupby(['gender', 'status'])['sl_no'].count()
counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```

###### Higher ssc marks have a higher chance of placement

```python
# Box plot of ssc percentage and placement status
sns.boxplot(x='status', y='ssc_p', data=df)
plt.show()
```

```python
# Bar plot of placement and ssc board
counts = df.groupby(['ssc_b', 'status'])['sl_no'].count()
counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```

###### Higher hsc marks have a higher chance of placement

```python
# Box plot of hsc percentage and placement status
sns.boxplot(x='status', y='hsc_p', data=df)
plt.show()
```

###### Other boards have a higher chance of placement

```python
# Bar plot of placement and hsc board
counts = df.groupby(['hsc_b', 'status'])['sl_no'].count()
counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```

###### Commerce stream has a higher chance of placement

```python
# Bar plot of placement and hsc stream
counts = df.groupby(['hsc_s', 'status'])['sl_no'].count()
counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```

###### Commerce and Management branched have a higher chance of placement

```python
# Bar plot of placement and degree branch
counts = df.groupby(['degree_t', 'status'])['sl_no'].count()
counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```

###### Higher degree marks have a higher chance of placement

```python
# Box plot of degree percentage and placement status
sns.boxplot(x='status', y='degree_p', data=df)
plt.show()
```

###### Higher etest scores have a slightly higher chance of placement

```python
# Box plot of etest percentage and placement status
sns.boxplot(x='status', y='etest_p', data=df)
plt.show()
```

###### Marketing and Finance has a higher chance of placement

```python
# Bar plot of placement and specialisation
gender_counts = df.groupby(['specialisation', 'status'])['sl_no'].count()
gender_counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```

###### Work experience has a significantly higher chance of placement

```python
# Bar plot of placement and workex
gender_counts = df.groupby(['workex', 'status'])['sl_no'].count()
gender_counts.plot(kind='bar')
plt.ylabel('Number of Students Placed')
plt.show()
```
