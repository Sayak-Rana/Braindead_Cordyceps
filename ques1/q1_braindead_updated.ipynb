```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/Placement_Data_Full_Class.csv")
```

```python
# Checking for missing values
df.isnull().sum()

# Descriptive Statistics
df.describe()
```

# **What are the factors affecting the salary of a student?**

###### Correlation matrix for continuous variables

```python
# Correlation matrix
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

###### Science branches have highest salary

```python
# Box plot of salaries by degree branch
sns.boxplot(x="degree_t", y="salary", data=df)
plt.show()
```

###### Degree percentage does not affect salary

```python
# Scatter plot of salaries by degree prcentage
correlation_value = df['degree_p'].corr(df['salary'])
plt.scatter(x="degree_p", y="salary", data=df)
plt.title(f"Correlation: {correlation_value}")
plt.xlabel("degree_p")
plt.ylabel("Salary")
plt.show()
```

###### Ssc Board does not affect salary

```python
# Box plot of salaries by ssc board
sns.boxplot(x="ssc_b", y="salary", data=df)
plt.show()
```

###### Higher Ssc percentage leads to slightly higher salary

```python
# Scatter plot of salaries by ssc percentage
correlation_value = df['ssc_p'].corr(df['salary'])
plt.scatter(x="ssc_p", y="salary", data=df)
plt.title(f"Correlation: {correlation_value}")
plt.xlabel("ssc_p")
plt.ylabel("Salary")
plt.show()
```

###### Hsc Board does not affect salary

```python
# Box plot of salaries by hsc board
sns.boxplot(x="hsc_b", y="salary", data=df)
plt.show()
```

###### Hsc stream does not affect salary, although arts stream has slightly lower salary than the others

```python
# Box plot of salaries by hsc stream
sns.boxplot(x="hsc_s", y="salary", data=df)
plt.show()
```

# **What are the factors affecting placement?**

###### Males have a slightly higher chance of placement

```python
# Bar plot of placement and gender
counts = df.groupby(['gender', 'status'])['sl_no'].count()
total_counts = df.groupby('gender')['sl_no'].count()
data = np.array([counts[1]/total_counts[0], counts[3]/total_counts[1]])
plt.bar(total_counts.index, data)
plt.ylabel('Percentage of Students Placed')
plt.show()
```

###### Higher ssc marks have a higher chance of placement

```python
# Box plot of ssc percentage and placement status
sns.boxplot(x='status', y='ssc_p', data=df)
plt.show()
```

###### Ssc board does not affect placement

```python
# Bar plot of placement and ssc board
counts = df.groupby(['ssc_b', 'status'])['sl_no'].count()
total_counts = df.groupby('ssc_b')['sl_no'].count()
data = np.array([counts[1]/total_counts[0], counts[3]/total_counts[1]])
plt.bar(total_counts.index, data)
plt.ylabel('Percentage of Students Placed')
plt.show()
```

###### Higher hsc marks have a higher chance of placement

```python
# Box plot of hsc percentage and placement status
sns.boxplot(x='status', y='hsc_p', data=df)
plt.show()
```
